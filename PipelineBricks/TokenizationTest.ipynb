```python
from Utilities.xml_parser import parse_xml

xml_tree = parse_xml("WO2012080152A1.xml")
description = xml_tree[0]["lexisnexis-patent-document"]["description"]["p"]
desc_text = []
for d in description:
    k = d.keys()
    if "#text" in k:
        desc_text.append(d["#text"])
desc_text
```

```python
desc_text = " ".join(desc_text)
# desc_text

tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
tokens = tokenizer.tokenize(desc_text)
print(len(tokens))

"""
BERT doesn't allow more than 512 tokens
an established solution is to cut out the middle of the text
"""
```

```python
claims = xml_tree[0]["lexisnexis-patent-document"]["claims"]["claim"]
print(claims[1]["claim-text"]["#text"])
```

```python
claim_list = []
for c in claims:
    k = c.keys()
    if "claim-text" in k:
        claim_text = c["claim-text"]
        if type(claim_text) != str:
            if "#text" in claim_text.keys():
                sub_text = claim_text["#text"]
                claim_list.append(sub_text)
        else:
            claim_list.append(claim_text)

claim_text = " ".join(claim_list)
tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
tokens = tokenizer.tokenize(claim_text)
print(len(tokens))
```

```python
abstract = xml_tree[0]["lexisnexis-patent-document"]["abstract"]

if type(abstract) == list:
    for element in abstract:
        if "@lang" in element.keys():
            if element["@lang"] == "eng":
                abstract_text = element["p"]
            else:
                break

print(abstract_text)

tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
tokens = tokenizer.tokenize(abstract_text)
print(tokens)
print(len(tokens))
```
